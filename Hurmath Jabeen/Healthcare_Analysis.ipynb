```python
#Patients with Hypertension and diabetes from medications
meds_hypdia = meds[meds['indication'].isin(['hypertension','diabetes','type2_diabetes'])].reset_index(drop=True)
meds_hypdia.index = meds_hypdia.index + 1
print(meds_hypdia)
```

```python
#Patients with Hypertension and diabetes from diagnoses
diagnoses_hypdia = diagnoses[diagnoses['primary_diagnosis'].isin(['hypertension','type2_diabetes'])].reset_index(drop=True)
diagnoses_hypdia.index = diagnoses_hypdia.index + 1
#print(diagnoses_hypdia)
print(diagnoses_hypdia['primary_diagnosis'].unique())
```

```python
#Patients with Hypertension and diabetes from labs
#labs['test_name'].unique()
labs_hypdia = labs[labs['test_name'].isin(['glucose_fasting','HbA1c','potassium','sodium'])].reset_index(drop=True)
labs_hypdia.index = labs_hypdia.index + 1
print(labs_hypdia)
print(labs_hypdia['test_name'].unique())
```

```python
from functools import reduce
#merging patients, meds_hypdia, diagnoses_hypdia, labs_hypdia data to get centralized view
dfs = [patients, meds_hypdia, diagnoses_hypdia, labs_hypdia]
merged_df = reduce(lambda left, right: pd.merge(left, right, on='patient_id', how='inner'), dfs)
merged_df.info()
```

```python
#readmitted patients for hypertension and diabetes
#All visits for hypertension and diabetes
patients_readmission = diagnoses[diagnoses['primary_diagnosis'].isin(['hypertension', 'type2_diabetes'])].copy()
patients_readmission
```

```python
# Sorting by patient and date to get multiple visits
patients_readmission['visit_date'] = pd.to_datetime(patients_readmission['visit_date'])
patients_readmission = patients_readmission.sort_values(['patient_id', 'visit_date'])
patients_readmission
```

```python
# Count visits per patient
visit_counts = patients_readmission.groupby('patient_id').size().reset_index(name='visit_count')
visit_counts.info()
```

```python
# patients with multiple visits-readmitted
readmitted_patients = visit_counts[visit_counts['visit_count'] > 1]

# Merge to get full details
readmitted_details = patients_readmission[patients_readmission['patient_id'].isin(readmitted_patients['patient_id'])]

print("Patients readmitted for hypertension/diabetes:")
print(f"Total unique patients readmitted: {len(readmitted_patients)}")
readmitted_details
```

```python
#CORRELATION HEATMAP
#numeric columns for correlation
numeric_cols = ['age', 'bmi', 'systolic_bp', 'diastolic_bp', 'heart_rate', 'temperature_f', 'charlson_index', 'dx_hypertension', 'dx_type2_diabetes', 'dx_hyperlipidemia']

corr_matrix = patients[numeric_cols].corr()


sns.heatmap(corr_matrix, annot=True, linewidths=0.5)
plt.title('Correlation Heatmap of Patient Health Indicators')
plt.tight_layout()
plt.show()
```

```python
#AGE DISTRIBUTION HISTOGRAM
plt.figure(figsize=(10, 6))
sns.histplot(patients['age'], bins=30, kde=True, color='skyblue')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#BOXPLOT: AGE BY HYPERTENSION STATUS
plt.figure(figsize=(8, 6))
sns.boxplot(x='dx_hypertension', y='age', data=patients)
plt.title('Age Distribution by Hypertension Status')
plt.xlabel('Hypertension (0=No, 1=Yes)')
plt.ylabel('Age')
plt.show()
```

```python
#TREND: AGE VS SYSTOLIC BP
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='systolic_bp', data=patients, alpha=0.3, color='purple')
sns.regplot(x='age', y='systolic_bp', data=patients, scatter=False)
plt.title('Trend: Age vs Systolic Blood Pressure')
plt.xlabel('Age')
plt.ylabel('Systolic Blood Pressure')
plt.show()
```

```python
#Insights
#1.Age distribution of patients
print("AGE PATTERNS")
print(f"Average age: {merged_df['age'].mean():.1f} years")
print(f"Age range: {merged_df['age'].min()} - {merged_df['age'].max()} years")

#Age groups
merged_df['age_group'] = '18-40'
merged_df.loc[merged_df['age'] > 40, 'age_group'] = '41-60'
merged_df.loc[merged_df['age'] > 60, 'age_group'] = '61-80'
merged_df.loc[merged_df['age'] > 80, 'age_group'] = '80+'
print("\nAge group distribution:")
print(merged_df['age_group'].value_counts())
```

```python
#2.Count patients with any diabetes
diabetes_patients = merged_df[(merged_df['dx_type1_diabetes'] == 1) | (merged_df['dx_type2_diabetes'] == 1)]
print(f"Total diabetic patients: {len(diabetes_patients)} out of {len(patients)}")
```

```python
#3.Count patients with any diabetes
hypertension_patients = merged_df[(merged_df['dx_hypertension'] == 1) | (merged_df['dx_hypertension'] == 1)]
print(f"Total diabetic patients: {len(hypertension_patients)} out of {len(patients)}")
```

```python
#4.Blood pressure patterns
print("BLOOD PRESSURE PATTERNS")
uncontrolled_htn = (merged_df['systolic_bp'] >= 140).sum()
print(f"Patients with uncontrolled BP (>=140): {uncontrolled_htn} ({uncontrolled_htn/len(merged_df)*100:.1f}%) out of {len(patients)}")
```

```python
#5.Compare BMI between hypertensive and non-hypertensive patients
hyper_bmi = merged_df[merged_df['dx_hypertension'] == 1]['bmi'].mean()
non_hyper_bmi = merged_df[merged_df['dx_hypertension'] == 0]['bmi'].mean()
print(f"Hypertensive patients BMI: {hyper_bmi:.1f}, Non-hypertensive: {non_hyper_bmi:.1f}")
```

```python
#6.Compare BMI between diabetes and non-diabetes patients
dia_bmi = merged_df[merged_df['dx_type2_diabetes'] == 1]['bmi'].mean()
non_dia_bmi = merged_df[merged_df['dx_type2_diabetes'] == 0]['bmi'].mean()
print(f"Diabetes patients BMI: {dia_bmi:.1f}, Non-diabetes: {non_dia_bmi:.1f}")
```

```python
#7.Average HbA1c levels for readmitted vs non-readmitted
readmitted_hba1c = merged_df[merged_df['patient_id'].isin(readmitted_patients['patient_id']) & (merged_df['test_name'] == 'HbA1c')]['value'].mean()
non_readmitted_hba1c = merged_df[~merged_df['patient_id'].isin(readmitted_patients['patient_id']) & (merged_df['test_name'] == 'HbA1c')]['value'].mean()
print(f"1. Readmitted patients have higher HbA1c ({readmitted_hba1c:.1f}% vs {non_readmitted_hba1c:.1f}%)")
```

```python
#8.Potassium levels
readmitted_k = merged_df[merged_df['patient_id'].isin(readmitted_patients['patient_id']) & (merged_df['test_name'] == 'potassium')]['value'].mean()
non_readmitted_k = merged_df[~merged_df['patient_id'].isin(readmitted_patients['patient_id']) & (merged_df['test_name'] == 'potassium')]['value'].mean()
print(f"4. Potassium levels differ (readmitted: {readmitted_k:.2f} vs {non_readmitted_k:.2f} mEq/L)")
```

```python
#9.Compare lab results by medication adherence
good_adherence = merged_df[merged_df['adherence_pct'] > 85]
poor_adherence = merged_df[merged_df['adherence_pct'] < 65]
print("Good adherence abnormal labs:", good_adherence['is_abnormal'].mean())
print("Poor adherence abnormal labs:", poor_adherence['is_abnormal'].mean())
```

```python
#10.Blood pressure patterns in diabetic patients
print("Average BP for diabetic vs non-diabetic patients")
patients['has_diabetes'] = patients['dx_type2_diabetes'] | patients['dx_type1_diabetes']
bp_by_diabetes = patients.groupby('has_diabetes')[['systolic_bp', 'diastolic_bp']].mean()
print(bp_by_diabetes)
```
